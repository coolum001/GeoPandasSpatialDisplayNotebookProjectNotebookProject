```python
def blank_axes(ax):
    """
    blank_axes:  blank the extraneous spines and tick marks for an axes

    Input:
    ax:  a matplotlib Axes object

    Output: None
    """


    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
#    ax.tick_params(labelbottom='off', labeltop='off', labelleft='off', labelright='off' ,\
#                    bottom='off', top='off', left='off', right='off' )
    ax.tick_params(labelbottom=False, labeltop=False, labelleft=False, labelright=False ,\
                    bottom=False, top=False, left=False, right=False )    
#end blank_axes
```

```python
place_name = "94 Banner St, London , United Kingdom"
lat, lon = ox.utils.geocode(place_name)

graph1 = ox.graph_from_address(place_name, network_type='drive')
g_node1, g_edge1 = ox.save_load.graph_to_gdfs(graph1)
gdf = ox.projection.project_gdf(g_edge1)


fig = plt.figure(figsize=(10,10) )
ax = fig.add_subplot(1, 1, 1, projection=ccrs.UTM(gdf.crs['zone'], southern_hemisphere=False), )
ax.set_aspect('equal')
ax.background_patch.set_facecolor('black')
fig.patch.set_facecolor('black')


# plot roads
gdf.plot(ax=ax,  edgecolor='white', linewidth=1, )

# plot marker with lon / lat
home_lat, home_lon = lat, lon
ax.plot(home_lon, home_lat, marker='+', transform=ccrs.PlateCarree(), markersize=20, alpha=0.4,  )

# plot same marker in raw UTM coords
x1, y1 =ax.projection.transform_point(home_lon, home_lat, src_crs=ccrs.PlateCarree()) 
ax.plot(x1, y1, marker='o', color='red')

plt.tight_layout()
ax.set_title(place_name,  pad=50, color='grey')

# plotting gdfb2 in same projection as Axes, so no transform parameter needed
gdfb = ox.buildings_from_point((home_lat, home_lon), distance=1000)
gdfb2 = ox.projection.project_gdf(gdfb)
gdfb2.plot(ax=ax, color='darkgrey',  )

ax.patch.set_facecolor('black')
fig.patch.set_facecolor('black')

plt.show()
```

```python
place_name = "94 Banner St, London , United Kingdom"
lat, lon = ox.utils.geocode(place_name)

graph1 = ox.graph_from_address(place_name, network_type='drive')
g_node1, g_edge1 = ox.save_load.graph_to_gdfs(graph1)
gdf = ox.projection.project_gdf(g_edge1)


fig = plt.figure(figsize=(10,10) )
ax = fig.add_subplot(1, 1, 1, projection=ccrs.UTM(gdf.crs['zone'], southern_hemisphere=False), )
ax.set_aspect('equal')
ax.background_patch.set_facecolor('black')
fig.patch.set_facecolor('black')


# plot roads
gdf.plot(ax=ax,  edgecolor='white', linewidth=1, )

# plot marker with lon / lat
home_lat, home_lon = lat, lon
ax.plot(home_lon, home_lat, marker='+', transform=ccrs.PlateCarree(), markersize=20, alpha=0.4,  )

# plot same marker in raw UTM coords
x1, y1 =ax.projection.transform_point(home_lon, home_lat, src_crs=ccrs.PlateCarree()) 
ax.plot(x1, y1, marker='o', color='red')

plt.tight_layout()
ax.set_title(place_name,  pad=50, color='grey')

xmin, xmax, ymin, ymax = ax.get_extent()
bar_offset = [0.0, 0.0, 0.02]
bar_x0 = xmin+ (xmax-xmin)*bar_offset[0]
bar_y0 = ymin + (ymax-ymin)*bar_offset[1]

text_x0 = bar_x0
text_y0 = ymin + (ymax-ymin)*bar_offset[2]
bar_tickmark = 100 # metres
bar_ticks = 5
bar_alpha = 0.3
bar_color = ['grey', 'red']

# draw a scale bar that is a set of colored line segments (bar_ticks of these), bar_tickmarks long
for i in range(bar_ticks):
     #  horizontal scale bar, y stays constant
    end_y, end_x = bar_y0, (bar_x0 + bar_tickmark)
    # capstyle must be set so line segments end square
  
    ax.plot([bar_x0, end_x], [bar_y0, end_y], 
            color=bar_color[i%2], 
            linewidth=20, 
             solid_capstyle='butt', 
            alpha = bar_alpha)
    # start of next bar is end of last bar
    bar_x0 = end_x
    bar_y0 = end_y
#end for

# highlight text with white background
buffer = [patheffects.withStroke(linewidth=3, foreground="grey")]
# Plot the scalebar label
units = 'mtr'

t0 = ax.text(text_x0, text_y0, 
             str(bar_ticks*bar_tickmark) + ' ' + units, 
             horizontalalignment='left', 
             verticalalignment='bottom',
             path_effects=buffer, 
             zorder=2)

ax.spines['right'].set_visible(True)
ax.spines['top'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

ax.spines['bottom'].set_color('grey')
ax.spines['top'].set_color('grey')
ax.spines['left'].set_color('grey')
ax.spines['right'].set_color('grey')

plt.show()
```
